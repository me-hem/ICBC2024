```python
import json
import matplotlib.pyplot as plt

# Specify the path to your JSON file
json_file_path = r"c:\Users\Umesh Bhatt\Desktop\EIP-3675-Analysis\Consensus Analysis\miningDistribution\1M_Pre_miningDistribution.json"

# Read miner data from the JSON file
with open(json_file_path, 'r') as json_file:
    miner_data = json.load(json_file)

# Select the first 10 miners
selected_miners = list(miner_data.keys())[:10]

# Create a dictionary to store the first 20 block numbers mined by each miner
mined_block_numbers = {}


# Iterate through the selected miners and extract their last 50 block numbers
for miner in selected_miners:
    mined_block_numbers[miner[:5]+".."] = miner_data[miner][-50:]
    

# Create a scatter plot for each miner

#plt.figure(figsize=(20, 10))
for miner, block_numbers in mined_block_numbers.items():
    plt.scatter(block_numbers, [miner] * len(block_numbers))
    

plt.xticks() 
plt.yticks() 

plt.xlabel('Block Number')
plt.ylabel('Miner\'s Public Key')
plt.grid(True)
plt.tight_layout()
#plt.show()
```
